```python
# load data.csv into pandas

import pandas as pd

data = pd.read_csv('data.csv', delimiter=';')

# convert target to numeric
# Dropout -> 0, Enrolled, Graduate -> 1
data['Target'] = data['Target'].map({'Dropout': 0, 'Enrolled': 1, 'Graduate': 1})

X = data.drop('Target', axis=1)
y = data['Target']

X
```

```python
# Nacionality -> Speaks Portuguese

# List of Portuguese-speaking countries
portuguese_speaking = [1, 21, 22, 24, 25, 26, 41]

# Create a new column 'Speaks_Portuguese'
X['Speaks_Portuguese'] = X['Nacionality'].apply(lambda x: 1 if x in portuguese_speaking else 0)

# Drop 'Nacionality' column
X.drop(columns=['Nacionality'], inplace=True)
```

```python
# Drop gender to avoid potential discrimination
X.drop(columns=['Gender'], inplace=True)
```

```python
# credited, enrolled -> Success Rate = credited / enrolled
X['Success_Rate_1st'] = X['Curricular units 1st sem (credited)'] / X['Curricular units 1st sem (enrolled)']
X.drop(columns=['Curricular units 1st sem (credited)'], inplace=True)
X.drop(columns=['Curricular units 1st sem (enrolled)'], inplace=True)
X.drop(columns=['Curricular units 1st sem (evaluations)'], inplace=True)
X.drop(columns=['Curricular units 1st sem (approved)'], inplace=True)
X.drop(columns=['Curricular units 1st sem (without evaluations)'], inplace=True)

X['Success_Rate_2nd'] = X['Curricular units 2nd sem (credited)'] / X['Curricular units 2nd sem (enrolled)']
X.drop(columns=['Curricular units 2nd sem (credited)'], inplace=True)
X.drop(columns=['Curricular units 2nd sem (enrolled)'], inplace=True)
X.drop(columns=['Curricular units 2nd sem (evaluations)'], inplace=True)
X.drop(columns=['Curricular units 2nd sem (approved)'], inplace=True)
X.drop(columns=['Curricular units 2nd sem (without evaluations)'], inplace=True)
```
